```python
#----Prompt Chaining(Prompt Zinciri)
#KarmaÅŸÄ±k iÅŸlemleri adÄ±m adÄ±m yapmayÄ± saÄŸlar. 
# 1.adÄ±m:veri toplama 2.adÄ±m : Ã¶zetleme 3.adÄ±m : Analiz etme
#Prompt Chaining yaparken her adÄ±mÄ±n Ã§Ä±ktÄ±sÄ±nÄ± bir sonraki adÄ±mÄ±n girdisi olarak kullanÄ±lÄ±r

from openai import OpenAI
client = OpenAI(
    base_url = "https://openrouter.ai/api/v1",
    api_key = ""
)
#Bilgi alma completionu
completion1 = client.chat.completions.create(
    model ="",
    messages = [
        {
            "role":"user",
            # 1.adÄ±m: Bilgileri toplama
            "content":(
                "BÃ¼yÃ¼k dil modelleri hakkÄ±nda 5 Ã¶nemli bilgi ver"
            )
        }
    ]
)

#1.adÄ±mÄ±n cevabÄ± alÄ±nÄ±r
cevap1 = completion1.choices[0].message.content
print("\n====1.AdÄ±m : Bilgi Toplama====\n")
print(cevap1)

#2.adÄ±m : Ã–zetleme completionu
completion2 = client.chat.completions.create(
    model = "",
    messages = [
        {
            "role":"user",
            "content":(
                f"AÅŸaÄŸÄ±daki bilgileri kÄ±sa bir paragrafla Ã¶zetle:\n\n{cevap1}"
            )
        }
    ]
)
#2.adÄ±mÄ±n cevabÄ± alÄ±nÄ±r.
cevap2 = completion2.choices[0].message.content
print("\n====2.AdÄ±m: Ã–zetleme====\n")
print(cevap2)

#3.adÄ±m: Analiz etme completionu
completion3 = client.chat.completions.create(
    model = "",
    messages = [
        {
            "role":"user",
            "content":(
                f"Bu Ã¶zet Ã¼zerinden yapay zekanÄ±n modern dÃ¼nyaya etkilerini analiz et\n\nÃ–zet:\n{cevap2}"
            )
        }
    ]
)
#3.AdÄ±mÄ±n cevabÄ±
cevap3 = completion3.choices[0].message.content
print("\n====3.AdÄ±m: Analiz SonuÃ§larÄ±====\n")
print(cevap3)
```
