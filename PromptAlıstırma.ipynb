```python
#Prompt modele verilen talimatlar dizisidir.Bu sayfada prompt alıştırmaları yapılacaktır.

#----ZERO SHOT PROMPTİNG örnek 1

#Gerekli kütüphaneler eklenir
from openai import OpenAI
#API işlemleri için
client = OpenAI(
    #openrouter'ın url adresi
    base_url="https://openrouter.ai/api/v1",
    #oluşturduğun api
    api_key ="",
)
#Modele sohbet tamamlama isteği göndermek için
completion = client.chat.completions.create(
    model =" ", #OpenRouter dan alınan model
    #modele gönderilen sohbetin listesi verilir
    messages = [
        {
            "role" : "user",
            "content" : "Bilgisayar Mühendislerinin geleceği hakkında kısa bir bilgi ver"
        }
            
    ]
)
#Modelin verdiği cevabı ekrana bastırmak için
print(completion.choices[0].message.content)
```

```python
#----Prompt Chaining(Prompt Zinciri)
#Karmaşık işlemleri adım adım yapmayı sağlar. 
# 1.adım:veri toplama 2.adım : özetleme 3.adım : Analiz etme
#Prompt Chaining yaparken her adımın çıktısını bir sonraki adımın girdisi olarak kullanılır

from openai import OpenAI
client = OpenAI(
    base_url = "https://openrouter.ai/api/v1",
    api_key = ""
)
#Bilgi alma completionu
completion1 = client.chat.completions.create(
    model ="",
    messages = [
        {
            "role":"user",
            # 1.adım: Bilgileri toplama
            "content":(
                "Büyük dil modelleri hakkında 5 önemli bilgi ver"
            )
        }
    ]
)

#1.adımın cevabı alınır
cevap1 = completion1.choices[0].message.content
print("\n====1.Adım : Bilgi Toplama====\n")
print(cevap1)

#2.adım : Özetleme completionu
completion2 = client.chat.completions.create(
    model = "",
    messages = [
        {
            "role":"user",
            "content":(
                f"Aşağıdaki bilgileri kısa bir paragrafla özetle:\n\n{cevap1}"
            )
        }
    ]
)
#2.adımın cevabı alınır.
cevap2 = completion2.choices[0].message.content
print("\n====2.Adım: Özetleme====\n")
print(cevap2)

#3.adım: Analiz etme completionu
completion3 = client.chat.completions.create(
    model = "",
    messages = [
        {
            "role":"user",
            "content":(
                f"Bu özet üzerinden yapay zekanın modern dünyaya etkilerini analiz et\n\nÖzet:\n{cevap2}"
            )
        }
    ]
)
#3.Adımın cevabı
cevap3 = completion3.choices[0].message.content
print("\n====3.Adım: Analiz Sonuçları====\n")
print(cevap3)
```

```python
#Prompt chaining örnek 2
from openai import OpenAI
client = OpenAI(
    base_url = "https://openrouter.ai/api/v1",
    api_key = ""
)
#1.adım: Bilgi alma
completion1 = client.chat.completions.create(
    model = "",
    messages = [
        {
            "role":"user",
            "content":(
                " İstanbul hakkında 100 kelimelik 5 bilgi ver\n\n"
            )
        }
    ]
)
#Bilgiyi alma
cevap1 = completion1.choices[0].message.content
print("\n====1.Adım : Bilgi Toplama====\n")
print(cevap1)

# 2.ADIM:Özetleme
completion2 = client.chat.completions.create(
    model ="",
    messages = [
        {
         "role":"user",
         "content":(
             f"Aşağıdaki bilgileri kısa bir paragrafla özetle\n\n{cevap1}"
         )
        }
    ]
)
cevap2 = completion2.choices[0].message.content
print("\n====2.Adım: Özetleme====\n")
print(cevap2)

#Özetin üzerinde analiz yapma
completion3 = client.chat.completions.create(
    model = "",
    messages = [
        {
            "role":"user",
            "content":(
                f"Yapmış olduğun özet üzerinde 1-2 cümlelik analiz yap\n\n{cevap2}"
            )
        }
    ]
)
cevap3=completion3.choices[0].message.content
print("\n====3.Adım: Analiz====\n")
print(cevap3)
```
